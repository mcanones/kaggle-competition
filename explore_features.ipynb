### Highly Correlated Features

```python
sns.heatmap(df.corr())
```

x, y and z are highly correlated with price. It's bad having highly correlated features in the dataset because they can distortion the resutls from the prediction. We could consider discarding these columns for the analysis (try and see what happens).
